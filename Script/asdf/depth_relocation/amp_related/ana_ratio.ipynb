```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = json.load( open( "./amp_ratio.json" ) )
```

```python
data["201803181800A"]
```

```python
# get the median value for each gmtid and each category
data_median={}
for gcmtid in data:
    data_median[gcmtid]={}
    for category in data[gcmtid]:
        data_median[gcmtid][category]=np.median(data[gcmtid][category])
```

```python
data_median
```

```python
pz_sz,pr_sr,pz_pr,sz_sr,sz_st,sr_st=[[] for i in range(6)]
for gcmtid in data_median:
    pz_sz.append(data_median[gcmtid]["pz_sz"])
    pr_sr.append(data_median[gcmtid]["pr_sr"])
    pz_pr.append(data_median[gcmtid]["pz_pr"])
    sz_sr.append(data_median[gcmtid]["sz_sr"])
    sz_st.append(data_median[gcmtid]["sz_st"])
    sr_st.append(data_median[gcmtid]["sr_st"])
```
