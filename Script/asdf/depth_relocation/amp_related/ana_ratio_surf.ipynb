```python
import json
import numpy as np
import matplotlib.pyplot as plt
import obspy
%matplotlib inline
```

```python
data = json.load( open( "./amp_ratio.json" ) )
```

```python
# get the median value for each gmtid and each category
data_median={}
for gcmtid in data:
    data_median[gcmtid]={}
    for category in data[gcmtid]:
        data_median[gcmtid][category]=np.median(data[gcmtid][category])
```

```python
data_median
```

```python
surfz_pz,surfz_sz,surfr_pr,surfr_sr,surfz_surfr=[[] for i in range(5)]
for gcmtid in data_median:
    surfz_pz.append(data_median[gcmtid]["surfz_pz"])
    surfz_sz.append(data_median[gcmtid]["surfz_sz"])
    surfr_pr.append(data_median[gcmtid]["surfr_pr"])
    surfr_sr.append(data_median[gcmtid]["surfr_sr"])
    surfz_surfr.append(data_median[gcmtid]["surfz_surfr"])
```

```python
depth_mapper={}
for event in events:
    depth_mapper[event.resource_id.id.split("/")[-2]]=event.origins[0].depth/1000
```

```python
# get the median value for each gmtid and each category
data_median={}
for gcmtid in data:
    if(depth_mapper[gcmtid]>150):
        continue
    data_median[gcmtid]={}
    for category in data[gcmtid]:
        data_median[gcmtid][category]=np.median(data[gcmtid][category])
```
